```python
# api for current covid data by noneil
import json
import pandas as pd
import requests
import matplotlib.pyplot as plt

url= 'https://api.covid19api.com/summary'
r = requests.get(url)
# <Response [200]> means success in getting the data
# 404 is bad responce
r
```

```python
# look at dict keys available
json = r.json()
json.keys() 
#countries is a list of dicts so turn into a df for use
```

```python
# now lets look at the data in a dataframe for each contry
#df = pd.Dataframe(json) # for plotting
df = pd.DataFrame(json['Countries']) # for looking at summarys
# remove the # below for full df
#pd.set_option("max_rows", None)
df
```

```python
# create object for plotting
fig = plt.figure()
ax = fig.add_subplot()
x = df.NewDeaths
ax.hist(x)
```
